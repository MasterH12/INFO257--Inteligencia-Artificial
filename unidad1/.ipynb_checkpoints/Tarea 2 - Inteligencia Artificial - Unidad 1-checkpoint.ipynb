```python
import pandas as pd
df = pd.read_csv("datos/SDSS-DR14.csv")
df.head()
```

```python
#Tomé prestada la función que cuenta cuanta cantidad de cada clase hay, y le hice una modificación para que funcionara en el
#dataframe
#SECCIÓN FUNCIONES e IMPORTACIONES DE LIBRERÍAS
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

def counts(rows, columna, modo, valor):
    """Counts the number of each type of example in a dataset."""
    counts = {}  # a dictionary of label -> count.
    rows = rows[columna]
    for row in rows:
        # in our dataset format, the label is always the last column
        label = row
        if(modo == "atipico<="):
            if label <= valor:
                if label not in counts:
                    counts[label] = 0
                counts[label] += 1
        elif(modo == "atipico>="):
            if label >= valor:
                if label not in counts:
                    counts[label] = 0
                counts[label] += 1
        else:
            if label not in counts:
                counts[label] = 0
            counts[label] += 1
    return counts
def limpiar(matriz, columna, modo, valor):
    counts = {}  # a dictionary of label -> count.
    if(modo==">="):
        matriz = matriz[matriz[columna]>=valor]
    elif(modo=="<="):
        matriz = matriz[matriz[columna]<=valor]
    return matriz
def distribucion(data, nombre):
    dist = norm(loc=np.mean(data), scale=np.std(data))
    x = np.linspace(dist.ppf(0.001),
                    dist.ppf(0.999), 100)
    plt.hist(data, normed=True, label=nombre)
    plt.plot(x, dist.pdf(x), "r-", label=nombre)
    plt.legend()
    return 0
#Para ver si los parámetros impactarán de forma significativa las predicciones de el arbol de decisión, se revisa
#si los datos son variados, y se verifica si se diferencian los datos entre clases
print("clases: ", counts(df,"class", "",""))
#Separamos los datos para mejor visualización en los gráficos
STAR = df[df["class"] == "STAR"]
GALAXY = df[df["class"] == "GALAXY"]
QSO = df[df["class"] == "QSO"]

#Balanceamos un poco QSO:
#QSO_aumentado = QSO.sample(4152, replace = True)
#print(counts(df_QSO_aumentado, "class", ""))
#df = pd.concat([STAR, GALAXY, QSO_aumentado], axis= 0)
#print("clases después: ", counts(STAR,"class", ""))

#Ahora está lo suficientemente balanceado, ya que ninguna clase sobrepasa a otra en una proporción mayor a 1
#print(STAR)
#VAMOS A REVISAR TODAS LAS COLUMNAS:
```

```python
#RA

plt.scatter(x=df["class"], y=df["ra"], c="blue") #Son casi todas las observaciones iguales
plt.xlabel("class")
plt.ylabel("ra")
plt.show()

distribucion(STAR["ra"],"star")
distribucion(GALAXY["ra"],"galaxy")
distribucion(QSO["ra"],"qso")
```

```python
#dec
plt.scatter(x=df["class"], y=df["dec"], c="blue") #parece que estos datos si podrían hacer la diferencia
#plt.scatter(x=df["objid"][df["class"]=="GALAXY"], y=df["ra"][df["class"]=="GALAXY"], c="green")
plt.xlabel("class")
plt.ylabel("dec")
plt.show()

distribucion(STAR["dec"],"star")
distribucion(GALAXY["dec"],"galaxy")
distribucion(QSO["dec"],"qso")
```

```python
#u
#Según los gráficos, también puede aportar, se pueden encontrar y descartar datos atípicos
#Se puede ver que se necesita una cierta limpieza de datos atípicos en la clase STAR
plt.scatter(x=df["class"], y=df["u"], c="blue") 
plt.xlabel("class")
plt.ylabel("u")
plt.show()
#procedemos a limpiar:
STAR = limpiar(STAR, "u", ">=", 14.5)

distribucion(STAR["u"],"star")
distribucion(GALAXY["u"],"galaxy")
distribucion(QSO["u"],"qso")
```

```python
#g
plt.scatter(x=df["class"], y=df["g"], c="blue") #parece que estos datos si podrían hacer la diferencia
plt.xlabel("class")
plt.ylabel("g")
plt.show()

print(counts(STAR, "g", "atipico<=", 14)) #Se puede observar que hay un dato atípico por debajo de este valor, por lo tanto se elimina
print(counts(QSO, "g", "atipico<=", 16)) #Y también hay un dato atípico bajo 16 en QSO
STAR = limpiar(STAR, "g", ">=", 14)
QSO= limpiar(QSO, "g", ">=", 16)

distribucion(STAR["g"],"star")
distribucion(GALAXY["g"],"galaxy")
distribucion(QSO["g"],"qso")
```
